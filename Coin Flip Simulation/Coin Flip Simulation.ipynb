```python
## Display a coin flip simulation

from random import randrange

# Initiate varibales in order to be able to find average amount of flips
COUNT = 0
TOTAL_FLIPS = 0

# Read the number of iteration from the user
n = int(input('Enter the number of iterations: '))
print()

# Create a simulation
for k in range(n):
    # In order not to face any indexing error, do find the first two flip results of each iteration outside of the while loop 
    first = randrange(1, 3)
    second = randrange(1, 3)
    # Initiate a list that contains flip results of each iteration
    my_list = [first, second]
    # Initiate a variable that shows amount of flips after each iteration 
    flips = 2
    # Initiate a variable that helps to find when to stop flipping
    index = 1
    # Initiate a variable that will contain flipping result of each iteration as a string
    my_string = ''
    # Compute the other reslts of flipping
    while not (len(my_list) >= 3 and my_list[index-2] == my_list[index - 1] == my_list[index]):
        m = randrange(1, 3)
        my_list.append(m)
        index += 1
        flips += 1
        # Replace numbers with the special characters
        for i, j in enumerate(my_list):
            if j == 1:
                my_list[i] = 'H'
            elif j == 2: 
                my_list[i] = 'T'
    COUNT += 1
    TOTAL_FLIPS += flips
    # Concatenate the special characters into one string after each iteration
    for element in my_list:
        my_string = my_string + element + '  '
    # Display the results after each iteration
    print(str(k+1)+'. ', my_string, '(%d flips)' %flips)
    print()
    
# Display the summary results
print('On average,', TOTAL_FLIPS / COUNT, 'flips were needed.')
```
